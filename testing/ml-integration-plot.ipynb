```python
import json
```

```python
with open("Ag_P_S_analysis.archive.json", encoding="utf-8") as f:
    data = json.load(f)
plotly_fig_json = data['data']['results'][0]['figures'][0]['figure']
```

```python
plotly_fig_json
```

```python
import plotly.io as pio
fig = pio.from_json(json.dumps(plotly_fig_json))
fig.update_layout(
    xaxis=dict(
        range=[-40, 0], dtick=10,
        tickfont=dict(size=16), title_font=dict(size=18),
        showline=True, linewidth=1, linecolor='white', mirror=True,
        ticks='outside', ticklen=8, tickwidth=2, tickcolor='black',
    ),
    yaxis=dict(
        range=[-40, 0], dtick=10, scaleanchor='x', scaleratio=1,
        tickfont=dict(size=16), title_font=dict(size=18),
        showline=True, linewidth=1, linecolor='white', mirror=True,
        ticks='outside', ticklen=8, tickwidth=2, tickcolor='black',
    ),
    width=600,
    height=600,
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)',
)
fig.show()
fig.write_image("ml-integration-plot.svg", scale=3)
fig.write_image("ml-integration-plot.png", scale=3)  # 1800x1800 px @ ~300 DPI
```
